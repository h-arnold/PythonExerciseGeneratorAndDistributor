## Exercise 1 — Adding numbers, not strings

**Working code:**
```python
a = "5"
b = "10"
print(a + b)
```

**Task:** Cast a and b to integers so it prints 15.

**Expected output:**
```
15
```


````python
a = "5"
b = "10"
# Convert from strings to integers
a = int(a)
b = int(b)
print(a + b)
````
