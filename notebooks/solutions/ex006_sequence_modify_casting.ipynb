## Exercise 1 — Adding numbers, not strings

**Working code:**
```python
a = "5"
b = "10"
print(a + b)
```

**Task:** Cast a and b to integers so it prints 15.

**Expected output:**
```
15
```


````python
a = "5"
b = "10"
# Convert from strings to integers
a = int(a)
b = int(b)
print(a + b)
````

## Exercise 10 — The Receipt

**Working code:**
```python

p2 = input()

```

**Task:** Cast inputs to floats, sum them, and print 'Total: 30.0' for inputs 10 and 20, showing prompts 'Enter item 1:' and 'Enter item 2:' in the example.

**Expected output example:**
```
Enter item 1: [Input: 10]

```


````python
print("Enter item 1:")
p1 = input()
print("Enter item 2:")
p2 = input()
# Cast inputs to float and sum to get the numeric total
p1 = float(p1)
p2 = float(p2)
print("Total: " + str(p1 + p2))
````
