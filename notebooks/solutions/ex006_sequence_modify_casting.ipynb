## Exercise 7 — Solution

## Exercise 10 — Solution

```python
# Exercise 10 — SOLUTION
print("Enter number of items:")
num_items = int(input())
print("Enter price per item:")
price_per_item = float(input())
print("Enter tax percent (e.g. 5):")
tax_percent = float(input())
total = num_items * price_per_item * (1 + tax_percent / 100)
print(f"Total: {total:.1f}")
```
