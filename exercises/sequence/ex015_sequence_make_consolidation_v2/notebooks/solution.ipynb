## Exercise 1 — Party Supply Calculator

Calculate full goodie bags and leftover sweets using floor division and modulus.

```
# Exercise 1 — Party Supply Calculator
# Follow the instructions step by step.

total_sweets = 50
sweets_per_bag = 7
full_bags = total_sweets // sweets_per_bag
sweets_left = total_sweets % sweets_per_bag
print(f"You can make {full_bags} full bags with {sweets_left} sweets left over.")
```

## Exercise 2 — Paint Coverage Calculator

Calculate wall area and how many cans of paint are needed.

```
# Exercise 2 — Paint Coverage Calculator

width = 5
height = 3
coverage_per_can = 12
area = width * height
cans_needed = (area + coverage_per_can - 1) // coverage_per_can
paint_left = (cans_needed * coverage_per_can) - area
print(f"Wall area: {area} square metres")
print(f"You need {cans_needed} cans; {paint_left} square metres of paint will be unused.")
```

## Exercise 3 — Pet Age Calculator

Convert human years to dog years by multiplying by 7.

```
# Exercise 3 — Pet Age Calculator

print("How old are you in human years?")
human_years = int(input())
dog_years = human_years * 7
print(f"If you were a dog, you would be {dog_years} years old.")
```

## Exercise 4 — Pizza Sharing

Calculate slices per person and leftover slices.

```
# Exercise 4 — Pizza Sharing

print("Number of pizzas:")
pizzas = int(input())
print("Number of people:")
people = int(input())
total_slices = pizzas * 8
slices_per_person = total_slices // people
leftover_slices = total_slices % people
print(f"If you ordered {pizzas} pizzas for {people} people:")
print(f"Each person gets {slices_per_person} slices, with {leftover_slices} slices left over.")
```

## Exercise 5 — Fuel Cost Calculator

Calculate total fuel cost and round to 2 decimal places.

```
# Exercise 5 — Fuel Cost Calculator

print("Litres of fuel:")
litres = float(input())
print("Price per litre in pounds:")
price_per_litre = float(input())
total_cost = litres * price_per_litre
print(f"{litres} litres at \u00a3{price_per_litre} per litre costs \u00a3{total_cost:.2f}.")
```

## Exercise 6 — Temperature Converter

Convert Celsius to Fahrenheit using the formula.

```
# Exercise 6 — Temperature Converter

print("Enter temperature in \u00b0C:")
celsius = int(input())
fahrenheit = celsius * 9 / 5 + 32
fahrenheit_rounded = round(fahrenheit, 1)
print(f"{celsius}\u00b0C is {fahrenheit_rounded}\u00b0F")
```

## Exercise 7 — Savings Goal

Calculate how many full weeks to reach a savings goal and the remainder.

```
# Exercise 7 — Savings Goal

print("What is your savings goal in pounds?")
goal = int(input())
print("How much can you save each week in pounds?")
weekly_savings = int(input())
full_weeks = goal // weekly_savings
left_to_save = goal % weekly_savings
print(f"Saving \u00a3{weekly_savings} per week towards a \u00a3{goal} goal:")
print(f"It will take {full_weeks} full weeks, with \u00a3{left_to_save} left to save.")
```

## Exercise 8 — Garden Area

Calculate garden area and the side length needed to double the area.

```
# Exercise 8 — Garden Area

print("Side length of the square garden in metres:")
side = int(input())
area = side ** 2
double_area_side = (area * 2) ** 0.5
double_area_side_rounded = round(double_area_side, 2)
print(f"A square garden with side {side}m has area {area} square metres.")
print(f"A garden with double the area would have side {double_area_side_rounded}m.")
```

## Exercise 9 — Bill Splitter with Tip

Calculate tip, total with tip, and per-person share.

```
# Exercise 9 — Bill Splitter with Tip

print("Total bill amount in \u00a3:")
bill = float(input())
print("Tip percentage (e.g. 10 for 10%):")
tip_percent = int(input())
print("Number of people sharing:")
people = int(input())
tip_amount = bill * tip_percent / 100
total_with_tip = bill + tip_amount
per_person = total_with_tip / people
per_person_rounded = round(per_person, 2)
print(f"Total bill: \u00a3{bill:.2f}")
print(f"Tip ({tip_percent}%): \u00a3{tip_amount:.2f}")
print(f"Total with tip: \u00a3{total_with_tip:.2f}")
print(f"Each of {people} people pays: \u00a3{per_person_rounded}")
```

## Exercise 10 — Classroom Supplies Order

Calculate packs of pencils and erasers needed, plus leftovers.

```
# Exercise 10 — Classroom Supplies Order

print("Number of students:")
students = int(input())
print("Pencils needed per student:")
pencils_per_student = int(input())
print("Erasers needed per student:")
erasers_per_student = int(input())

total_pencils = students * pencils_per_student
total_erasers = students * erasers_per_student

pencil_pack_size = 24
eraser_pack_size = 15

pencil_packs = (total_pencils + pencil_pack_size - 1) // pencil_pack_size
eraser_packs = (total_erasers + eraser_pack_size - 1) // eraser_pack_size

leftover_pencils = (pencil_packs * pencil_pack_size) - total_pencils
leftover_erasers = (eraser_packs * eraser_pack_size) - total_erasers

print(f"For {students} students:")
print(f"Each student needs {pencils_per_student} pencils and {erasers_per_student} erasers.")
print(f"Order {pencil_packs} packs of pencils ({pencil_pack_size} per pack) \u2014 {leftover_pencils} pencils will be left over.")
print(f"Order {eraser_packs} packs of erasers ({eraser_pack_size} per pack) \u2014 {leftover_erasers} erasers will be left over.")
```
