## Exercise 1 — Full groups only

**Expected output:**
```
Full groups: 6
```

**Task:** Fix the operator bug AND convert the `print` line to use an f-string instead of `+` concatenation.


````python
students = 25
group_size = 4
full_groups = students // group_size
print(f"Full groups: {full_groups}")
````

### What went wrong
The code used `/` (regular division), which gave `6.25` instead of `6`. Floor division `//` drops the decimal part because you cannot have part of a group. The `print` line also used `+` concatenation with `str()`, which was replaced with a cleaner f-string.

**Fix:** Changed `/` to `//` and converted `print("..." + str(x))` to `print(f"...{x}")`.


## Exercise 2 — Find the leftover

**Expected output:**
```
Leftover: 1
```

**Task:** Fix the operator bug AND convert the `print` line to use an f-string instead of `+` concatenation.


````python
sweets = 25
per_bag = 4
leftover = sweets % per_bag
print(f"Leftover: {leftover}")
````

### What went wrong
The code used `-` (subtraction), which gave `21` instead of `1`. Subtraction takes away a fixed number, but modulus `%` finds the remainder after making full groups. The `print` line also used `+ str()` concatenation, which was replaced with an f-string.

**Fix:** Changed `sweets - per_bag` to `sweets % per_bag` and converted the print line to use an f-string.


## Exercise 3 — Round an average

**Expected output:**
```
Average: 6.7
```

````python
total = 20
count = 3
average = total / count
average = round(average, 1)
print(f"Average: {average}")
````

### What went wrong
The average was printed directly without rounding, producing `6.666666666666667` instead of `6.7`. The `round()` function is needed to control how many decimal places are shown.

**Fix:** Added `average = round(average, 1)` before the print line to round to 1 decimal place.


## Exercise 4 — Teams from input

**Expected output:**
```
How many players? [⌨️: 23]
Full teams: 4
```

````python
players = int(input("How many players? "))
teams = players // 5
print(f"Full teams: {teams}")
````

### What went wrong
`input()` always returns a string, so `players // 5` caused a `TypeError` — you cannot use floor division on a string. The input must be converted to an integer first.

**Fix:** Wrapped `input(...)` inside `int(...)` so `players` becomes a number before the `//` calculation.


## Exercise 5 — Round money to 2 decimal places

**Expected output:**
```
Total: £3.77
```

````python
price = 1.257
qty = 3
total = price * qty
total = round(total, 2)
print(f"Total: £{total}")
````

### What went wrong
The `round()` call was correct, but the f-string printed `{price}` instead of `{total}`. The original price (`1.257`) was shown instead of the rounded total (`3.77`).

**Fix:** Changed `{price}` to `{total}` in the f-string so the rounded calculation is displayed.


## Exercise 6 — Split minutes

**Expected output:**
```
200 minutes is 3 hours and 20 minutes
```

````python
minutes = 200
hours = minutes // 60
minutes_left = minutes % 60
print(f"{minutes} minutes is {hours} hours and {minutes_left} minutes")
````

### What went wrong
Two bugs: `/` gave a decimal number of hours (`3.333...`) instead of whole hours, and `- 60` subtracted exactly 60 minutes rather than finding the remainder. The output showed `3.333... hours and 140 minutes`.

**Fix:** Changed `/` to `//` for whole hours and `- 60` to `% 60` for leftover minutes.


## Exercise 7 — Pence to pounds

**Expected output:**
```
Enter pence: [⌨️: 389]
389p is £3 and 89p
```

````python
pence = int(input("Enter pence: "))
pounds = pence // 100
leftover = pence % 100
print(f"{pence}p is £{pounds} and {leftover}p")
````

### What went wrong
`input()` returns a string, causing a `TypeError` on `pence // 100`. Also, `pence - 100` subtracts exactly 100 pence rather than finding the true remainder after pounds are taken out.

**Fix:** Wrapped `input(...)` in `int(...)` and changed `- 100` to `% 100` to properly split pence into pounds and leftover pence.


## Exercise 8 — Area of a square

**Expected output:**
```
Area of square: 5.5
```

````python
length = 2.345
area = length * length
rounded = round(area, 2)
print(f"Area of square: {rounded}")
````

### What went wrong
`round(area, 3)` rounded to 3 decimal places (`5.499`) instead of 2 (`5.5`). Also, the f-string printed `{area}` (the unrounded value) instead of `{rounded}`.

**Fix:** Changed `round(area, 3)` to `round(area, 2)` and changed `{area}` to `{rounded}` in the f-string.


## Exercise 9 — Apples into bags

**Expected output:**
```
Full bags: 9, Left over: 2
```

````python
apples = 47
per_bag = 5
bags = apples // per_bag
left = apples % per_bag
print(f"Full bags: {bags}, Left over: {left}")
````

### What went wrong
The print line was missing the `f` prefix, so `{bags}` and `{left}` were printed literally as text instead of being replaced with their values. Also, `/` was used instead of `//`, so `bags` was `9.4` instead of `9`.

**Fix:** Added `f` before the string and changed `/` to `//` for whole-number bag count.


## Exercise 10 — Mixed challenge

**Expected output:**
```
How many items? [⌨️: 29]
How many per box? [⌨️: 6]
Total cost? £ [⌨️: 10.99]
Full boxes: 4
Leftover items: 5
Cost per box: £2.75
```

````python
items = int(input("How many items? "))
box_size = int(input("How many per box? "))
total_cost = float(input("Total cost? £"))

boxes = items // box_size
leftover = items % box_size
cost_per_box = total_cost / boxes
cost_per_box = round(cost_per_box, 2)

print(f"Full boxes: {boxes}")
print(f"Leftover items: {leftover}")
print(f"Cost per box: £{cost_per_box}")
````

### What went wrong
Four bugs: (1) `items = input(...)` needed `int()` casting, (2) `box_size = input(...)` also needed `int()` casting, (3) `/` was used for box count instead of `//`, and (4) `-` was used for leftover instead of `%`. The program crashed with a `TypeError` before reaching the print lines.

**Fix:** Added `int()` to both `input()` calls, changed `/` to `//` for `boxes`, and changed `-` to `%` for `leftover`.
