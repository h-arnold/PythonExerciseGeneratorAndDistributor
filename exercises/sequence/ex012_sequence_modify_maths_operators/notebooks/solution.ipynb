## Exercise 1 — Full groups only

**Working code:**
```python
students = 29
group_size = 4
full_groups = students / group_size
print("Full groups: " + str(full_groups))
```

**Task:** Change the calculation so it uses floor division `//`. This gives only whole groups and drops the decimal part. We do this because you cannot have part of a student in a group.

**Expected output:**
```
Full groups: 7
```


````
# Exercise 1 — SOLUTION
students = 29
group_size = 4
full_groups = students // group_size
print("Full groups: " + str(full_groups))
````

## Exercise 10 — Mixed challenge

**Working code:**
```python
crayons = 26
per_box = 4
total_price = 10.99

full_boxes = crayons / per_box
leftover = crayons - per_box
price_per_box = total_price / full_boxes

print("Full boxes: " + str(full_boxes))
print("Leftover crayons: " + str(leftover))
print("Price per box: £" + str(price_per_box))
```

**Task:** Fix all three calculations: use floor division `//` for full boxes, modulus `%` for leftover crayons, and `round(..., 2)` for price per box.

**Expected output:**
```
Full boxes: 6
Leftover crayons: 2
Price per box: £1.83
```


````
# Exercise 10 — SOLUTION
crayons = 26
per_box = 4
total_price = 10.99

full_boxes = crayons // per_box
leftover = crayons % per_box
price_per_box = total_price / full_boxes
price_per_box = round(price_per_box, 2)

print("Full boxes: " + str(full_boxes))
print("Leftover crayons: " + str(leftover))
print("Price per box: £" + str(price_per_box))
````
