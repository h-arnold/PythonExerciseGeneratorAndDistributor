## Exercise 1
Create a short program that stores the words `Oakwood`, `Coding`, and `Club` in separate variables.
Print exactly one line:

`Welcome to Oakwood Coding Club!`

Use the variables to build the message.


```python
# Exercise 1
school = "Oakwood"
subject = "Coding"
group = "Club"
message = "Welcome to " + school + " " + subject + " " + group + "!"
print(message)
```

## Exercise 2
A snack box contains 4 muffins and each muffin costs 2 pounds.
Write a program that:
- stores the snack name, quantity, and price in variables
- works out the total cost
- prints exactly these two lines:

`Snack box: muffins`
`Total cost: 8 pounds`


```python
# Exercise 2
snack_name = "muffins"
quantity = 4
price = 2
total_cost = quantity * price
print("Snack box: " + snack_name)
print("Total cost: " + str(total_cost) + " pounds")
```

## Exercise 3
Write a program that asks the user for their name and favourite hobby.
Then print exactly one line in this style:

`Hello Aisha! Your hobby is drawing.`

Use `print()` before each `input()` so the prompts appear in the output.

## Exercise 4
Write a program that asks for the number of books read in one week and the number of weeks.
Convert both answers to integers, work out the total books read, and print exactly:

`Books read altogether: 18`

Example: if the user enters `6` and `3`, the final line should be `Books read altogether: 18`.


```python
# Exercise 4
print("Books read in one week:")
books_per_week = int(input())
print("Number of weeks:")
weeks = int(input())
total_books = books_per_week * weeks
print("Books read altogether: " + str(total_books))
```

## Exercise 5
Write a program that asks for the distance of one walk in kilometres and the number of walks completed.
Convert the distance to a float and the number of walks to an integer.
Work out the total distance and print exactly:

`Total distance: 7.5 km`

Example: if the user enters `2.5` and `3`, the final line should be `Total distance: 7.5 km`.


```python
# Exercise 5
print("Distance for one walk in km:")
distance_km = float(input())
print("Number of walks:")
walk_count = int(input())
total_distance = distance_km * walk_count
print("Total distance: " + str(total_distance) + " km")
```
