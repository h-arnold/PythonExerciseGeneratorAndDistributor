## New comparison operators
These two operators are new:

| Operator | Meaning | True example |
|----------|---------|--------------|
| `<=` | less than or equal to | `5 <= 5` is `True` |
| `>=` | greater than or equal to | `5 >= 5` is `True` |

So `total >= 50` is `True` when `total` is 50 **or more**.

## A quick word on constants
A **constant** is a value that should not change while the program runs. In Python we write constant names in UPPER_CASE (for example `FREE_DELIVERY_THRESHOLD = 50`) and place them near the top of the code. Storing fixed values like thresholds in constants makes your code easier to read and easier to change.

In the first three exercises you will simply *refactor* working code to use constants. From exercise 4 onwards the code already uses constants — your job is to change the constant's value and/or the operator, as well as the messages.

## Exercise 1 — Use constants for the delivery thresholds

**Task:**

- Introduce two constants: `FREE_DELIVERY_THRESHOLD = 50` and `STANDARD_DELIVERY_THRESHOLD = 20`.
- Use them in the `if` and `elif` conditions instead of the numbers.

The program should behave exactly the same as before.

**Expected output:**
```

```


````python
total_input = input("Enter your total spend: ")
total = int(total_input)
FREE_DELIVERY_THRESHOLD = 50
STANDARD_DELIVERY_THRESHOLD = 20
if total >= FREE_DELIVERY_THRESHOLD:
    print(f"Free delivery! Your total is £{total}")
elif total >= STANDARD_DELIVERY_THRESHOLD:
    print(f"Standard delivery for £{total} order")
else:
    print(f"Small order: £{total}")
````

## Exercise 9 — Grade bands with three elifs (change constants + rounding)

**Task:**

- Change the `GRADE_A` constant from `90` to `85` and the `GRADE_B` constant from `75` to `70`.
- Update the messages so the bands read A (for `>= 85`), B (for `>= 70`), C (for `>= 60`), D (for `>= 50`) and F otherwise.

The C, D and F thresholds stay the same.

**Expected output:**
```
Score 1: [⌨️: 80]
Score 2: [⌨️: 82]
Score 3: [⌨️: 78]
Grade B: 80.0
```


````python
s1_input = input("Score 1: ")
s1 = int(s1_input)
s2_input = input("Score 2: ")
s2 = int(s2_input)
s3_input = input("Score 3: ")
s3 = int(s3_input)
average = round((s1 + s2 + s3) / 3, 1)
GRADE_A = 85
GRADE_B = 70
GRADE_C = 60
GRADE_D = 50
if average >= GRADE_A:
    print(f"Grade A: {average}")
elif average >= GRADE_B:
    print(f"Grade B: {average}")
elif average >= GRADE_C:
    print(f"Grade C: {average}")
elif average >= GRADE_D:
    print(f"Grade D: {average}")
else:
    print(f"Grade F: {average}")
````

## Exercise 10 — Order sizing with multiple elifs (change formula + a constant)

**Task:**

- This is the hardest one! Add a delivery fee to the cost.
- After `cost = base * qty`, add two lines: `delivery = cost // 10` and `cost = cost + delivery`.
- Then change the `BULK_ORDER` constant from `200` to `250`.

Keep all the f-string messages the same.

**Expected output:**
```
Enter base price: [⌨️: 30]
Enter quantity: [⌨️: 4]
Bulk order, cost £132
```


````python
base_input = input("Enter base price: ")
base = int(base_input)
qty_input = input("Enter quantity: ")
qty = int(qty_input)
cost = base * qty
delivery = cost // 10
cost = cost + delivery
SMALL_ORDER = 20
MEDIUM_ORDER = 50
LARGE_ORDER = 100
BULK_ORDER = 250
if cost <= SMALL_ORDER:
    print(f"Small order, cost £{cost}")
elif cost <= MEDIUM_ORDER:
    print(f"Medium order, cost £{cost}")
elif cost <= LARGE_ORDER:
    print(f"Large order, cost £{cost}")
elif cost <= BULK_ORDER:
    print(f"Bulk order, cost £{cost}")
else:
    print(f"Wholesale order, cost £{cost}")
````
