# Selection Modify Basics

## Goal
Welcome to **selection** — making decisions in code! In these exercises you will practise modifying existing programs that use `if` statements.

By the end of this notebook you will be able to:
- Change what a program checks using `==`, `!=`, `<`, and `>`
- Change what a program prints when a condition is true
- Predict how changing a comparison affects the program's behaviour

## Comparison Operators
These are the four operators you will use:

| Operator | Meaning |
|----------|---------|
| `==` | equal to |
| `!=` | not equal to |
| `<` | less than |
| `>` | greater than |

## How the `if` statement works

```python
if condition:
    # this code runs only if the condition is True
    print("Something happens!")
```

The indented code underneath the `if` line only runs when the condition is **true**.

## Worked Example

Suppose you have this working program:

```python
colour = input("What is your favourite colour? ")
if colour == "blue":
    print("Blue is a nice colour!")
```

**Task:** Change it so the program says "Blue is the best!" when the user types "blue".

**Solution:** You only need to change the `print` message:

```python
colour = input("What is your favourite colour? ")
if colour == "blue":
    print("Blue is the best!")
```

**Expected output** (when the user types `blue`):
```
Blue is the best!
```
