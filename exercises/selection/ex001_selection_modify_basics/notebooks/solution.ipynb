## Exercise 2 — Change the temperature threshold

```
# Exercise 2 — SOLUTION
# Change 30 to 25
temp_input = input("What is the temperature in Celsius? ")
temp = int(temp_input)
if temp > 25:
    print("It's hot today!")
```

## Exercise 3 — Use "not equal to"

```
# Exercise 3 — SOLUTION
# Change == to != and update the message
num_input = input("Enter a number: ")
num = int(num_input)
if num != 10:
    print("That's not ten!")
```
